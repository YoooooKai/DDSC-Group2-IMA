```python
# !pip install scikit-learn or !pip install --user scikit-learn

import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 

# Only works on Jupyter/Anaconda
%matplotlib inline 

data = pd.read_csv("machine_data_for_student_4.csv", index_col = 0, encoding="ISO-8859-1")
data.head()
```

```python
print (data.dtypes)
```

```python
# Use the Kolmogorov-Smirnov test
from scipy.stats import kstest

for col in data.columns: # iteratre over each column
     # compare the column distribution to a normal distribution
    statistic, pval = kstest(data[col], "norm")
    print(f"column = {col}") 
    print(f"p value = {pval}")
    if pval < 0.05:
        print("Probably not normal")
    else:
        print("Probaly normal")
    print("\n")
```

```python
# create a correlation matrix rounding to one decimal point
correlation_matrix = data.corr().round(2)
# print a correlation heat map

plt.figure (figsize=(16,10))

sns.heatmap(data=correlation_matrix, annot=True)
```

```python
data.rename(columns={'Target_1':'status'},inplace = True)
data.head()
```

```python
# separate the target (y) from the features (x)
y = data.status
x = data.drop(['status'], axis=1)
```

```python
# split data into training and test
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

# print the shapes to check everything is OK
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Try the LogisticRegression

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression as LR
# fit a decision tree to the training data
clf = LR()
clf_fit = clf.fit(x_train, y_train)
```

```python
# Check the performance score again
# get the accuracy score of the test data
clf_fit.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'class_weight': ['balanced', None],
                     'C': [1, 10, 100],
                     'solver': ['liblinear', 'newton-cg']}
                   ]

scores = ['recall', 'f1_macro']

for score in scores:
    print("# Tuning hyperparameters for %s" % score)
    print("\n")
    clf = GridSearchCV(LR(), tuned_parameters, cv=5,
                       scoring= score)
    clf.fit(x_train, y_train)
    print("Best parameters set found on the training set:")
    print(clf.best_params_)
    print("\n")
```

```python
# Check
from sklearn.model_selection import cross_val_score
cross_val_score(LR(C=1, class_weight='balanced',solver='liblinear'), 
                x_train, y_train, cv=5).mean()
```
